```python
df_test=df.drop('Country', axis=1)
df_test=df_test.drop('Year', axis=1)
df_test
```

```python
abs_threshold = 0.6  # Valore assoluto soglia

correlations = abs(df_test.corr()['Life_exp']) < abs_threshold
selected_columns = correlations[correlations].index.tolist()
selected_columns
```

```python
#elimino le colonne con correlazione in valore assoluto minore di 0.65
df_finale=df_test.drop(columns=['infant deaths',
 'percentage expenditure',
 'Hepatitis B',
 'Measles ',
 ' BMI ',
 'under-five deaths ',
 'Polio',
 'Diphtheria ',
 ' thinness  1-19 years',
 ' thinness 5-9 years'])
df_finale
```

```python
X = df_test.drop(columns=['infant deaths',
 'percentage expenditure',
 'Hepatitis B',
 'Measles ',
 ' BMI ',
 'under-five deaths ',
 'Polio',
 'Diphtheria ',
 ' thinness  1-19 years',
 ' thinness 5-9 years','Life_exp'])
y = df['Life_exp']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 667
                                                    )
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('R2_score: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```

```python
import numpy as np
import matplotlib.pyplot as plt
length = y_pred.shape[0]
k = np.linspace(0,length,length)

plt.figure(figsize=(10,7))
plt.plot(k,y_test, label="Real")
plt.plot(k,y_pred, label="Previsione")
plt.legend(loc=2)
```
